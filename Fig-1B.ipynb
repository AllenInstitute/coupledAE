```python
sns.set_style("darkgrid")
def ax_style(ax):
    ax.set_facecolor('#EEEEEE')
    xlims = [-3,3]
    ylims = [-3,3]
    ax.set_xlim(xlims[0],xlims[1])
    ax.set_ylim(ylims[0],ylims[1])
    ax.set_xticks(np.linspace(xlims[0],xlims[1],num=6))
    ax.set_yticks(np.linspace(ylims[0],ylims[1],num=6))
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])
    return

save_path = '/Users/fruity/Dropbox/AllenInstitute/CellTypes/doc/NeurIPS 2019/camready/'
names = ['Fig-1B i','Fig-1B ii','Fig-1B iii','Fig-1B iv']
ind = np.random.randint(0,60000,5000)
for i,f in enumerate(fname):
    for j in range(2):
        X = sio.loadmat(pth+f,squeeze_me=True)
        plt.figure(figsize=(4, 4))
        plt.scatter(X['z_train_'+str(j)][ind,0], X['z_train_'+str(j)][ind,1], s=1, c=X['train_lbl'][ind],marker='.',cmap='tab10')
        ax = plt.gca()
        ax_style(ax)
        plt.savefig(save_path+names[i]+' '+str(j)+'.pdf',dpi=300, bbox_inches = 'tight')
        plt.show()
```
